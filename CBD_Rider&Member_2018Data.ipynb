```python
import pandas as pd
```

```python
jan_mar_df = pd.read_csv('raw_data/CitiBikeDailyRidershipAndMembershipData/data-Jan-Mar-2018.csv')
jan_mar_df.head()
```

```python
apr_jun_df = pd.read_csv('raw_data/CitiBikeDailyRidershipAndMembershipData/data-Apr-Jun-2018.csv')
apr_jun_df.head()
```

```python
jul_sep_df = pd.read_csv('raw_data/CitiBikeDailyRidershipAndMembershipData/data-Jul-Sep-2018.csv')
jul_sep_df.head()
```

```python
oct_dec_df = pd.read_csv('raw_data/CitiBikeDailyRidershipAndMembershipData/data-Oct-Dec-2018.csv')
oct_dec_df.head()
```

```python
riders_members_data = pd.concat([jan_mar_df,apr_jun_df,jul_sep_df,oct_dec_df])
riders_members_data.head()
```
